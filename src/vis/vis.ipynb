```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
DATA_PATH = "../../data/experiments/"
```

```python
y= [215, 122, 5.92, 6.14, 6.01, 5.2]
y2= [7.6, 8.6, 8.7, 8.1, 8.3, 8.1]
x_values = ["1", "0.25", "0.20", "0.15", "0.05", "0.01"]
x=np.arange(len(x_values))

fig, ax1 = plt.subplots(layout='constrained', figsize=[6.4*2, 4.8*2])
ax2 = ax1.twinx()


ax1.bar(x-0.2, y, 0.4, label='PercentError', color='green')
ax1.set_ylabel('PercentError', color='green')
ax1.set_xlabel('b-value')
ax1.set_xticks(x)
ax1.set_xticklabels(x_values)

ax2.bar(x+0.2, y2, 0.4,label='Time', color='red')
ax2.set_ylabel('Median Time (in s)', color='red')

plt.title("Error and Runtime for different b-values")
plt.show()
```
