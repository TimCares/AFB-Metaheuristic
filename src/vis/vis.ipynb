```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
DATA_PATH = "../../data/experiments/"
```

```python
def toRoundedThousandString(x):
    return str(int(round(x/1000))) + "k"
```

## eil101

### Benchmark

```python
data = pd.read_csv(f'{DATA_PATH}eil101 21_10_2023, 03:04.csv', header=None).values
data2 = pd.read_csv(f'{DATA_PATH}eil101 04_11_2023, 05:20.csv', header=None).values
plt.plot(data, color='blue', label='Benchmark')
plt.plot(data2[:1_000_000], color='red', label='TopN')
plt.plot(len(data)-1, data[-1], marker='o', color='blue', label=f'{round(data[-1][0])}')
plt.plot(1_000_000, data2[1_000_000], marker='o', color='red', label=f'{round(data2[1_000_000][0])}')
plt.axhline(y=629, color='black', linestyle='-.', label='629 (Optimum)')
plt.ylabel('Tour cost')
plt.xlabel('Iterations (In millions)')
plt.ylim(0)
plt.title("Cost of best tour over iterations (eil101)")
plt.legend()
plt.show()
```

### Join Top N (25%)

```python
data = pd.read_csv(f'{DATA_PATH}eil101 04_11_2023, 05:20.csv', header=None).values
plt.plot(data)
plt.plot(0, data[0], marker='o', color='red', label=f'{round(data[0][0])}')
plt.plot(1_000_000, data[1_000_000], marker='o', color='green', label=f'{round(data[1_000_000][0])}')
plt.axhline(y=629, color='black', linestyle='-.', label='629 (Optimum)')
plt.ylabel('Tour cost')
plt.xlabel('Iterations (In millions)')
plt.ylim(0)
plt.xlim((0, 1e6))
plt.title("Cost of best tour over iterations (eil101)")
plt.legend()
plt.show()
```

```python
data_cost = pd.read_csv(f'{DATA_PATH}eil101_birds.csv', header=None).values
data_time = pd.read_csv(f'{DATA_PATH}eil101_time.csv', header=None).values

n_birds = np.arange(5, 1001, 5)
n_birds_label = np.arange(5, 1001, 100)

fig, ax1 = plt.subplots(layout='constrained')
ax2 = ax1.twinx()

ax1.plot(n_birds, data_cost, label='Cost', color='green')
ax1.set_ylabel('Tour cost')
ax1.set_xlabel('Birds')
ax1.set_xticks(n_birds_label)
ax1.set_xticklabels(n_birds_label)

ax2.plot(n_birds, data_time, label='Time', color='red')
ax2.set_ylabel('Time (ms)')

plt.title("Effect of number of birds (eil101)")
plt.show()
```

```python
data_cost = pd.read_csv(f'{DATA_PATH}ali53529_10_2023, 05:16_birds.csv', header=None).values
data_time = pd.read_csv(f'{DATA_PATH}ali53529_10_2023, 05:16_time.csv', header=None).values
data_rounds = pd.read_csv(f'{DATA_PATH}ali53529_10_2023, 05:16_rounds.csv', header=None).values

n_birds = np.arange(5, 1001, 5)
n_birds_label = np.arange(5, 1011, 100)

fig, ax1 = plt.subplots(layout='constrained')
ax2 = ax1.twinx()


ax1.plot(n_birds, data_cost, label='Cost', color='green')
ax1.set_ylabel('Minimum tour cost', color='green')
ax1.set_xlabel('Birds')
ax1.set_xticks(n_birds_label)
ax1.set_xticklabels(n_birds_label)

ax2.plot(n_birds, data_time, label='Time', color='red')
ax2.set_ylabel('Time (ms)', color='red')
fig.legend(loc='upper left', bbox_to_anchor=(0.15, 0.9))

plt.title("Effect of number of birds\n(ali535)")
plt.show()
```

```python
fig, ax1 = plt.subplots(layout='constrained')


ax1.plot(n_birds, data_rounds)
ax1.set_ylabel('Moves per bird')
ax1.set_yscale(value='log')
ax1.set_xlabel('Birds')
ax1.set_xticks(n_birds_label)
ax1.set_xticklabels(n_birds_label)
ax1.plot(n_birds[0], data_rounds[0], marker='o', color='green', label=f'{data_rounds[0][0]+1}')
ax1.plot(n_birds[-1], data_rounds[-1], marker='o', color='red', label=f'{data_rounds[-1][0]+1}')
ax1.legend()

plt.title("Effect of number of birds on moves per bird\n(ali535)")
plt.show()
```
